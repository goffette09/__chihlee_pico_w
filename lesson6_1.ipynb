```python
for i in range(1,10):
    print(f"i = {i}")
    for j in range(1,10):
        print(f"j = {j}")
    print("=============")
```

```python
for i in range(1,10):
    print(f"i = {i}")
    for j in range(1,10):
        print(f"j={j}", end = ' ')
    print("=============")  # ===結果連在j=9之後 →下一段code修正
```

```python
for i in range(1,10):
    print(f"i = {i}")
    for j in range(1,10):
        print(f"j={j}", end = ' ')
    print() # 解決 " ===結果連在j=9之後 "的問題
    print("=============")
```

```python
for i in range(1,10):
    print(f"i = {i}")
    for j in range(1,10):
        print(f"j={j}", end = ' ')
    print("\n=============")  # \n換行；解決 " ===結果連在j=9之後 "的問題
```
